```python
%sql
USE CATALOG retail_sales_catalog;
USE SCHEMA bronze;

CREATE TABLE IF NOT EXISTS customers_raw (
  CustomerID INT,
  CustomerName STRING,
  Email STRING,
  City STRING,
  Address STRING,
  LastUpdated STRING
)
USING CSV
OPTIONS (
  path 's3://retaill-client/retaill-lakehouse/landing/customers_src_20042026100105.csv',
  header 'true'
);

CREATE TABLE IF NOT EXISTS customers_copy_raw (
  CustomerID INT,
  CustomerName STRING,
  Email STRING,
  City STRING,
  Address STRING,
  LastUpdated STRING
)
USING CSV
OPTIONS (
  path 's3://retaill-client/retaill-lakehouse/landing/customers_src_20042026100105 - Copy.csv',
  header 'true'
);

DROP TABLE IF EXISTS products_raw;
CREATE TABLE products_raw (
  ProductID INT,
  ProductName STRING,
  Category STRING,
  UnitPrice INT
)
USING CSV
OPTIONS (
  path 's3://retaill-client/retaill-lakehouse/landing/products_src_20042026100105.csv',
  header 'true'
);

DROP TABLE IF EXISTS stores_raw;
CREATE TABLE stores_raw (
  StoreID INT,
  StoreName STRING,
  Region STRING
)
USING CSV
OPTIONS (
  path  's3://retaill-client/retaill-lakehouse/landing/stores_src_20042026100107.csv',
  header 'true'
);

DROP TABLE IF EXISTS sales_raw;
CREATE TABLE sales_raw (
  TransactionID INT,
  CustomerID INT,
  ProductID INT,
  StoreID INT,
  Quantity INT,
  TxnDate DATE
)
USING CSV
OPTIONS (
  path 's3://retaill-client/retaill-lakehouse/landing/sales_transactions_src_20042026100107.csv',
  header 'true'
);

SELECT * FROM customers_raw LIMIT 10;
SELECT * FROM products_raw LIMIT 10;
SELECT * FROM stores_raw LIMIT 10;
SELECT * FROM sales_raw LIMIT 10;
```
